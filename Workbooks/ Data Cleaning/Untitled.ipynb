Great, looks like that has worked. Now let's try to remove punctuation in these columns as well before we try stemming and lemmatizing our data.


```python
nlp_df = nlp_df.applymap(lambda x: x.translate(x.maketrans('', '', string.punctuation)))
```

Great, looks like there is no punctuation in the document now, let's look at continuing to clean up this data. We shoudl remove stopwards and numbers at this point a so that we have an almost clean data set.

```python
sw = stopwords.words('english')
sw
```

```python
nlp_df['notes'] = nlp_df['notes'].apply(lambda x: ' '.join([word for word in x.split() if word not in (sw)]))
nlp_df['style'] = nlp_df['style'].apply(lambda x: ' '.join([word for word in x.split() if word not in (sw)]))
nlp_df['text'] = nlp_df['text'].apply(lambda x: ' '.join([word for word in x.split() if word not in (sw)]))


```

```python
nlp_df
```

```python
nlp_df = nlp_df.applymap(lambda x: x.translate(x.maketrans('', '', '0123456789')))
nlp_df
```

```python
nlp_df['combined'] = nlp_df['notes'] + ' ' + nlp_df['text']
```

```python
nlp_df
```
